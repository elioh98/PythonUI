```python
from tkinter import *
import os


def deletescreen(sc):
    sc.destroy()
def encrypt(text, s):
    result = ""



    for i in range(len(text)):
        char = text[i]


        if (char.isupper()):
            result += chr((ord(char) + s - 65) % 26 + 65)

        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
    return result

def signup_user():
    user_info = username.get()
    pass_info = password.get()
    f = open("elio.txt", "a")
    f.write("Username : "+user_info+" ")
    f.write("Password : "+encrypt(pass_info,4)+"\n")
    f.close()
    user_entry.delete(0, END)
    pass_entry.delete(0, END)
    Label(screen1, text="You have been successfully registered", fg="green").pack()

def success():
    global screen3
    screen3 = Toplevel(screen)
    screen3.title("Success")
    screen3.geometry("150x100")
    Label(screen3,bg="black",fg = "gold", text="Login Success").pack()
    Button(screen3,bg="black",fg = "cyan2", text="OK", command= lambda:deletescreen(screen3)).pack()


def wrongpass():
    global screen4
    screen4 = Toplevel(screen)
    screen4.title("Wrong Password")
    screen4.geometry("150x100")
    Label(screen4,bg="black",fg = "firebrick1", text="Password Error").pack()
    Button(screen4,bg="black",fg = "lawn green", text="OK", command=lambda:deletescreen(screen4)).pack()
def usererror():
    global screen5
    screen5 = Toplevel(screen)
    screen5.title("User Error")
    screen5.geometry("150x100")
    Label(screen5,bg="black",fg = "aquamarine", text="User Not Found").pack()
    Button(screen5,bg="black",fg = "IndianRed1", text="OK", command=lambda:deletescreen(screen5)).pack()
def loginverification():
    userc = user_verify.get()
    passc = pass_verify.get()
    user_entry1.delete(0,END)
    pass_entry1.delete(0,END)
    f = open("elio.txt","r")
    c = f.read().splitlines()
    found = False
    for i in c:
        if userc in i:
            found = True
            if encrypt(passc,4) in i:
                success()
                break
            else:
                wrongpass()
                break
    if not found:
        usererror()





def signup():
    global screen1
    screen1 = Toplevel(screen)
    screen1.geometry("600x400")
    screen1.title("Sign Up")
    global username
    global password
    global user_entry
    global pass_entry
    username = StringVar()
    password = StringVar()
    Label(screen1, bg="black", fg="SeaGreen1", text="Enter your username and password below ").pack()
    Label(screen1, text="").pack()
    Label(screen1, bg="black", fg="DarkOrchid1", text="Username").pack()
    user_entry = Entry(screen1, textvariable=username)
    user_entry.pack()
    Label(screen1, bg="black", fg="magenta2", text="Password").pack()
    pass_entry = Entry(screen1,show="*", textvariable=password)
    pass_entry.pack()
    Label(screen1, text="").pack()
    Button(screen1, text="Register", bg="black", fg="red", width="10", height="1", command=signup_user).pack()
def signin():
    global screen2
    screen2 = Toplevel(screen)
    screen2.geometry("600x400")
    screen2.title("Sign In")
    global user_verify
    global pass_verify
    global user_entry1
    global pass_entry1
    user_verify = StringVar()
    pass_verify = StringVar()
    Label(screen2, bg="black", fg="deep pink", text="Enter your username and password below to login ").pack()
    Label(screen2, text="").pack()
    Label(screen2, bg="black", fg="green2", text="Username").pack()
    user_entry1 = Entry(screen2, textvariable=user_verify)
    user_entry1.pack()
    Label(screen2 ,bg="black", fg="green2", text="Password").pack()
    pass_entry1 = Entry(screen2,show="*", textvariable=pass_verify)
    pass_entry1.pack()
    Label(screen2, text="").pack()
    Button(screen2, text="Sign in", bg="black", fg="red", width="10", height="1", command=loginverification).pack()



def main_screen():
    global screen
    screen = Tk()
    screen.geometry("600x400")
    screen.title("Elio's login system")
    Label(text="My first login project", bg="black", fg="blue", width="600", height="3", font=("Helvetica", 14)).pack()
    Label(text="").pack()
    Button(text="Sign in", bg="black", fg="yellow", height="2", width="40", command=signin).pack()
    Label(text="").pack()
    Button(text="Sign up", bg="black", fg="yellow", height="2", width="40", command=signup).pack()
    screen.mainloop()


main_screen()
```
